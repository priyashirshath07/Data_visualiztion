```python
#strip plot
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/content/Iris[1].csv")
print(df.head())
```

```python
ax=sns.stripplot(data=df,x="Species",y="SepalWidthCm")
ax.set(xlabel="Iris Species",ylabel="Sepal Width")
sns.set(style="whitegrid")
plt.title("Strip Plot in Python")
plt.show()
```

```python
ax=sns.swarmplot(data=df,x="Species",y="SepalWidthCm")
ax.set(xlabel="Iris Species",ylabel="Sepal Width")
sns.set(style="whitegrid")
plt.title("Swarm Plot in Python")
plt.show()
```

```python
#count plot
sns.countplot(x = "Species",data = df)
```

```python
sns.boxplot(x = "Species", y = "SepalWidthCm", data = df, hue = "Species")
```

```python
# violinplot"
sns.violinplot(x = "Species",y = "SepalWidthCm",data = df,hue = "Species")
```

```python
sns.set_style("whitegrid")
sns.boxplot(data = df)
plt.show()
```

```python
sns.set_style("whitegrid")
sns.violinplot(data = df)
plt.show()
```

```python
#pairplot
sns.pairplot(df,hue = "Species",height = 2.5)
```

```python
# piechart
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
df = pd.read_csv("/content/Iris[1].csv")
print(df.head())
df['Species'].value_counts()
```

```python
plt.figure(figsize=(10,10))
plt.style.use("Solarize_Light2")
df['Species'].value_counts().plot.pie()
plt.show()
```
